```python
#getting some basic information

from os import path
import pandas as pd
import numpy as np

path =  '/content/drive/MyDrive/Netflix Dataset.csv'
df = pd.read_csv(path)

print(df.shape)

print(df.size)

print(df.columns)

print(df.dtypes)

print(df.describe())

print(df.info())
```

```python
#Finding duplicate values and removing it

print(df.shape)
print(df.duplicated())
```

```python
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
#Finding null values and removing, if possible
print(df.isnull)
```

```python
print(df.isnull().sum())
```

```python
import seaborn as sb

sb.heatmap(df.isnull())
```

```python
#Number of Movies and TV shows
import seaborn as sb

df['Category'].value_counts()
```

```python
#Find all the types of instances

genre = (df['Type'].unique())
print(genre)
```

```python
unique_type = df['Type'].str.split(', ').explode().unique()
print(unique_type)
```

```python
#Find number of unique type
import matplotlib.pyplot as plt

unique_type_count = df['Type'].str.split(', ').explode().value_counts().copy()
print(unique_type_count)

unique_type_count.plot(kind='bar', figsize=(12, 6))
plt.title('Count of Each Content Type')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#Best movies and tv shows

best_movies = df[(df['Category'] == 'Movie') & (df['Rating']=='PG-13')]
display(best_movies[['Title', 'Type']])
print(best_movies.count())
```

```python
best_tv_shows = df[(df['Category'] == 'TV Show') & (df['Rating']=='TV-14')]
display(best_tv_shows[['Title', 'Type']])
print(best_tv_shows.count())
```

```python
#Movies and TV shows release in 2021

df['Date_N'] = pd.to_datetime(df['Release_Date'], format='mixed')
display(df.head())
```

```python
df['Year'] = df['Date_N'].dt.year
display(df.head())
```

```python
movies_2021 = df[(df['Category'] == 'Movie') & (df['Year'] == 2021)]
display(movies_2021[['Title', 'Type']].head(10))
print(movies_2021.count())
```

```python
tv_2021 = df[(df['Category'] == 'TV Show') & (df['Year'] == 2021)]
display(tv_2021[['Title', 'Type']].head(10))
print(tv_2021.count())
```

```python
#Number of Movies and TV shows release yearly

unique_years = df['Year'].unique()
print(unique_years)
print('Number of Years = ', len(unique_years))
```

```python
movies_df = df[df['Category'] == 'Movie'].copy()
display(movies_df.head())
```

```python
movies_per_year = movies_df.groupby('Year').size()
print(movies_per_year)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6), )
plt.scatter(movies_per_year.index, movies_per_year.values, color = 'blue')
plt.title('Number of Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

```python
Tv_df = df[df['Category'] == 'TV Show'].copy()
display(Tv_df.head())
```

```python
Tv_per_year = Tv_df.groupby('Year').size()
print(Tv_per_year)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.scatter(Tv_per_year.index, Tv_per_year.values, color = 'red')
plt.title('Number of TV Shows Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of TV Shows')
plt.grid(True)
plt.show()
```

```python
#Comparing yearly release of shows

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.scatter(movies_per_year.index, movies_per_year.values, label='Movies', color = 'blue')
plt.scatter(Tv_per_year.index, Tv_per_year.values, label='TV Shows', color = 'red')
plt.title('Number of Movies and TV Shows Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Content')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Display top 5 movies per type

movies_df = df[df['Category'] == 'Movie'].copy()
display(movies_df.head())
```

```python
type_movies = movies_df['Type'].str.split(', ').explode().value_counts()

print("Counts of each content type for movies:")
print(type_movies)
```

```python
expanded_movies_df = movies_df.assign(Type=movies_df['Type'].str.split(', ')).explode('Type').reset_index(drop=True)
display(expanded_movies_df.head())
print(expanded_movies_df.count())
```

```python
grouped_movies_by_type = expanded_movies_df.groupby('Type')
print(grouped_movies_by_type.groups.keys())
```

```python
top_5_movies_per_type = grouped_movies_by_type.head(5)
print("Top 5 movies per content type:")
display(top_5_movies_per_type[['Title', 'Type']])
```

```python
display(top_5_movies_per_type.sort_values(by='Type'))
print(top_5_movies_per_type.count())
```

```python
Tv_df = df[df['Category'] == 'TV Show'].copy()
print(Tv_df.head())
```

```python
type_Tv = Tv_df['Type'].str.split(', ').explode().value_counts()

print("Counts of each content type for TV shows:")
print(type_Tv)
```

```python
expanded_tv_df = Tv_df.assign(Type=Tv_df['Type'].str.split(', ')).explode('Type').reset_index(drop=True)
display(expanded_tv_df.head())
print(expanded_tv_df.count())
```

```python
grouped_tv_by_type = expanded_tv_df.groupby('Type')
print(grouped_tv_by_type.groups.keys())
```

```python
top_5_tv_per_type = grouped_tv_by_type.head(5)
print("Top 5 TV shows per content type:")
display(top_5_tv_per_type[['Title', 'Type']])
```

```python
display(top_5_tv_per_type.sort_values(by='Type'))
print(top_5_tv_per_type.count())
```
